```python
df.drop('index', inplace=True, axis=1)
df.head()
```

```python
df['Salary Estimate'].head()
```

```python
df['Salary Estimate'] = df['Salary Estimate'].str.replace(r'[$K]', '', regex=True)
df['Salary Estimate'].unique()
```

```python
df['Salary Estimate'] = df['Salary Estimate'].str.split('(').str[0]
df['Salary Estimate'].unique()
```

```python
df['min_salary'] = df['Salary Estimate'].str.split('-').str[0]
df['min_salary'] = df['min_salary'].astype(int)

df['max_salary'] = df['Salary Estimate'].str.split('-').str[1]
df['max_salary'] = df['max_salary'].astype(int)

df.head()
```

```python
df['avg_salary'] = (df['min_salary'] + df['max_salary']) // 2
df.head()
```

```python
df.sort_values(by='Rating', ascending=True)
```

```python
df['Rating'].dtype
```

```python
df['Rating'] = np.where(df['Rating'] == -1.0, 0.0, df['Rating'])
df.sort_values(by='Rating', ascending=True)
```

```python
df['Founded']
```

```python
df['company_age'] = np.where(df['Founded'] < 1, df['Founded'], 2023 - df['Founded'])
df['company_age'] = np.where(df['company_age'] == -1.0, 0.0, df['company_age'])
df['company_age']
```

```python
df['Job Description'][0]
```

```python
df['Job Description'][1].split('\n')
```

```python
df['python'] = df['Job Description'].str.contains('python', case=False).astype(int)
df.head()
```

```python
df['sql'] = df['Job Description'].str.contains('sql', case=False).astype(int)
df['aws'] = df['Job Description'].str.contains('aws', case=False).astype(int)
df['excel'] = df['Job Description'].str.contains('excel', case=False).astype(int)
df['spark'] = df['Job Description'].str.contains('spark', case=False).astype(int)
df['tableau'] = df['Job Description'].str.contains('tableau', case=False).astype(int)
df['big_data'] = df['Job Description'].str.contains('big_data', case=False).astype(int)
df['machine learning'] = df['Job Description'].str.contains('machine learning', case=False).astype(int)
df['powerbi'] = df['Job Description'].str.contains('powerbi', case=False).astype(int)

df.head()
```

```python
df['Job Title'].value_counts()
df['Job Title'].unique()
```

```python
def title_simplifier(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'machine learning'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    else:
        return 'na'
    
def seniority(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower() or 'sr.' in title.lower():
        return 'senior'
    elif 'jr' in title.lower() or 'jr.' in title.lower() or 'junior' in title.lower():
        return 'junior'
    else:
        return 'na'
```

```python
df['job_simp'] = df['Job Title'].apply(title_simplifier)
df['job_simp'].value_counts()
```

```python
df['seniority'] = df['Job Title'].apply(seniority)
df['seniority'].value_counts()
```

```python
df.drop(['Founded','Competitors'], axis=1,inplace=True)
```

```python
df = df.apply(lambda x: x.strip() if isinstance(x,str) else x)
```

```python
df.head()
```

```python
df['Company Name'] = df['Company Name'].str.split('\n').str[0]
df['Company Name'] = df['Company Name'].str.replace(r'\n.*', '', regex=True)

df.head()
```

```python
df.to_csv('Cleaned_DS_Jobs.csv', index=False)
```
